### Caso mais simples

Faça a separação dos dados para treinamento e teste. Reserve 20% dos dados para teste.

```python
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(Xpca, y, test_size=0.2, random_state=42)
```

Importe o regressor linear e utilize-o para gerar um modelo, então faça uma regressão com os dados de teste.

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train, y_train)
y_pred = reg.predict(X_test)
```

Importe as funções de erro quadrático médio (MSE) e o coeficiente de determinação (R²), então imprima o MSE e o R².

```python
from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse}")
print(f"R²: {r2}")
```

Então exiba o gráfico de dispersão dos dados de teste e a curva de regressão gerada.

```python
import matplotlib.pyplot as plt

plt.scatter(X_test, y_pred)
plt.title('Gráfico de Dispersão')
plt.xlabel('X teste')
plt.ylabel('y pred')
```

### Comparando Métodos de Redução de Dimensionalidade

Agora, a fim de comparar com outras técnicas, crie um método iterativo que realize as etapas de separação dos dados entre treinamento e teste, gere um modelo de regressão linear, faça a predição com base nos dados de testes e calcule o erro quadrático médio e o coeficiente de determinação para cada iteração. Então, calcule os valores mínimo, médio, máximo e desvio padrão para cada uma dessas duas métricas. Por fim, plote os histogramas (com 100 bins) das duas métricas. Use 10000 (dez mil) iterações.

#### PCA

```python
mses = []
r2s = []

for i in range(10000):
    pca = PCA(n_components=1, random_state=i)
    Xpca = pca.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(Xpca, y, test_size=0.2, random_state=i)
    
    reg = LinearRegression().fit(X_train, y_train)
    y_pred = reg.predict(X_test)
    
    mses.append(mean_squared_error(y_test, y_pred))
    r2s.append(r2_score(y_test, y_pred))

mse_min = min(mses)
r2_min = min(r2s)
print(f"Mínimo MSE: {mse_min}")
print(f"Mínimo R2: {r2_min}")

mse_max = max(mses)
r2_max = max(r2s)
print(f"Máximo MSE: {mse_max}")
print(f"Máximo R2: {r2_max}")

mse_mean = np.mean(mses)
r2_mean = np.mean(r2s)
print(f"Média MSE: {mse_mean}")
print(f"Média R2: {r2_mean}")

mse_std = np.std(mses)
r2_std = np.std(r2s)
print(f"Desvio padrão MSE: {mse_std}")
print(f"Desvio padrão R2: {r2_std}")
```

```python

fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the `bins` kwarg
axs[0].hist(mses, bins=100)
axs[1].hist(r2s, bins=100)
```

#### ICA

```python
mses = []
r2s = []

for i in range(10000):
    ica = FastICA(n_components=1, random_state=i)
    Xica = ica.fit_transform(X)
    
    X_train, X_test, y_train, y_test = train_test_split(Xica, y, test_size=0.2, random_state=i)
    
    reg = LinearRegression().fit(X_train, y_train)
    y_pred = reg.predict(X_test)
    
    mses.append(mean_squared_error(y_test, y_pred))
    r2s.append(r2_score(y_test, y_pred))

mse_min = min(mses)
r2_min = min(r2s)
print(f"Mínimo MSE: {mse_min}")
print(f"Mínimo R2: {r2_min}")

mse_max = max(mses)
r2_max = max(r2s)
print(f"Máximo MSE: {mse_max}")
print(f"Máximo R2: {r2_max}")

mse_mean = np.mean(mses)
r2_mean = np.mean(r2s)
print(f"Média MSE: {mse_mean}")
print(f"Média R2: {r2_mean}")

mse_std = np.std(mses)
r2_std = np.std(r2s)
print(f"Desvio padrão MSE: {mse_std}")
print(f"Desvio padrão R2: {r2_std}")
```

```python
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)

# We can set the number of bins with the `bins` kwarg
axs[0].hist(mses, bins=100)
axs[1].hist(r2s, bins=100)
```

### Comparando modelos de Regressão

Exiba o gráfico de dispersão dos dados de teste e a curva de regressão gerada em cada caso.

#### Least Squares

```python
X_train, X_test, y_train, y_test = train_test_split(Xpca, y, test_size=0.2, random_state=42)

reg = LinearRegression().fit(X_train, y_train)
y_pred_LS = reg.predict(X_test)

plt.scatter(X_test, y_test, color='grey')
plt.plot(X_test, y_pred_LS, color="green")

plt.show()
```

#### Ridge

```python
from sklearn import linear_model

regr = linear_model.Ridge()
regr.fit(X_train, y_train)
y_pred_Ri = regr.predict(X_test)

plt.scatter(X_test, y_test,  color='grey')
plt.plot(X_test, y_pred_Ri, color='blue')

plt.show()
```

#### Lasso

```python
regr = linear_model.Lasso()
regr.fit(X_train, y_train)
y_pred_La = regr.predict(X_test)

plt.scatter(X_test, y_test,  color='grey')
plt.plot(X_test, y_pred_La, color='yellow')

plt.show()
```

#### Lasso LARS

```python
regr = linear_model.LassoLars()
regr.fit(X_train, y_train)
y_pred_LL = regr.predict(X_test)

plt.scatter(X_test, y_test,  color='grey')
plt.plot(X_test, y_pred_LL, color='red', linewidth=2)

plt.show()
```

#### Bayesian Ridge

```python
regr = linear_model.BayesianRidge()
regr.fit(X_train, y_train)
y_pred_Ba = regr.predict(X_test)

plt.scatter(X_test, y_test,  color='grey')
plt.plot(X_test, y_pred_Ba, color='pink')

plt.show()
```

#### RANSAC

```python
regr = linear_model.RANSACRegressor()
regr.fit(X_train, y_train)
y_pred_Ra = regr.predict(X_test)

plt.scatter(X_test, y_test,  color='grey')
plt.plot(X_test, y_pred_Ra, color='orange')

plt.show()
```

#### Theil-Sen

```python
regr = linear_model.TheilSenRegressor()
regr.fit(X_train, y_train)
y_pred_TS = regr.predict(X_test)

plt.scatter(X_test, y_test,  color='grey')
plt.plot(X_test, y_pred_TS, color='purple')

plt.show()
```
